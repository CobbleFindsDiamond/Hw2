```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
import seaborn as sns
#
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, f1_score
import numpy as np
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from scipy.stats import ttest_rel
from sklearn.metrics import f1_score
data = pd.read_csv("/content/obesity-prediction-dataset/ObesityDataSet_raw_and_data_sinthetic.csv")
data.head
print(data.describe())
column_names = data.columns
print(column_names)
print(data.isnull().sum())
print(data.isna().sum()[lambda x: x > 0])
label_encoders = {}
for col in data.select_dtypes(include='object').columns.drop('NObeyesdad'):
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
print(data['NObeyesdad'].unique())
data['NObeyesdad'] = data['NObeyesdad'].map({'Insufficient_Weight': 0, 'Normal_Weight': 0, 'Overweight_Level_I': 0, 'Overweight_Level_II': 0, 'Obesity_Type_I': 1, 'Obesity_Type_II': 1, 'Obesity_Type_III': 1})
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
X = data.drop("NObeyesdad", axis=1)
Y = data["NObeyesdad"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=42)
model = DecisionTreeClassifier(random_state=42)
depths = [2, 4, None]
for i in depths:
    model = DecisionTreeClassifier(max_depth = i, random_state=42)
    model.fit(X_train, Y_train)
    plt.figure(figsize=(12, 6))
    plot_tree(model,  feature_names=X.columns, class_names = ['Not_Overweight', 'Overweight'], filled=True)
    plt.title("Decision Tree")
    plt.show()
acc = model.score(X_test, Y_test)
CopXOne = X_test.copy()
CopXTwo = X_test.copy()
CopXOne['Age'] = np.random.permutation(CopXOne['Age'])
CopXTwo['FCVC'] = np.random.permutation(CopXTwo['FCVC'])
accOne = acc - model.score(CopXOne, Y_test)
accTwo = acc - model.score(CopXTwo, Y_test)
print("Age drop: ", accOne)
print("FCVC drop: ", accTwo)
folds = [5,10,15]
models = {
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=40),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=42),
    "GradientBoost": GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42),
}
#orginally I was going to use KFold but later found out that StratifiedKFold was more consistent
#once I averaged it out
fOneALT = {name: [] for name in models}
for m in models:
    mods = models[m]
    for k in folds:
        kf = StratifiedKFold(n_splits=k, shuffle=True, random_state=40)
        acc = []
        precsion = []
        fOne = []
        #the actual spliting of the test occurs here
        for training, test in kf.split(X,Y):
             X_train = X.iloc[training]
             X_test = X.iloc[test]
             Y_train = Y.iloc[training]
             Y_test = Y.iloc[test]
             mods.fit(X_train,Y_train)
             Y_prediction = mods.predict(X_test)
             acc.append(accuracy_score(Y_test, Y_prediction))
             precsion.append(precision_score(Y_test, Y_prediction))
             fOne.append(f1_score(Y_test, Y_prediction))
             fOneALT[m].append(f1_score(Y_test, Y_prediction))
        print("For model", m)
        print("For ", k,"fold, ")
        print("accuracy: ", np.mean(acc))
        print("precsion: ", np.mean(precsion))
        print("fOne: ", np.mean(fOne))
```

#section 2
K-fold cross-validation is in essence a way to remedy the effects of spiting the data in training and testing data. When you split the data, you may unknowing bias your training towards that data. Additionally the data that you put into training cannot be used for testing, so every time you decide which is which you are losing on a potential training or testing candidate. K-fold cross-validation remedies this spiting the test into k folds or parts and then assigning one of those parts as the training and the rest for training while the rest is used for testing and repeats this k times selecting a different fold or part of the data to make into the training and the rest for testing. This can be seen in the code in the form of a for loop which runs through all these variations and then fitting them to the model. Afterwards we simply average them out in order to find the accuracy, F1, score, etc.


The performance metric really does not change much since the data I used is considered easy to predict. There are very clear trends that impact wether or not someone is obese. For example if the exercise regularly, there is a very high chance they are not obese. However I will explain what a difference in metrics mean anyway. To give a concise example, I will use accuracy and precision. Accuracy measures wether or not the model actually predicts the right things while precision measures the closeness of those true predictions. Recall to give another example measures the the 'right' decision so when the model predicts obese and not correctly. The scores may change depending on wether you are aiming for the model to be good at purely detecting wether something is 'NA' or when something is 'NA' and not 'NA'.

#Section 3
The models are all equally effective since they have everything the same. But if the models were different, a confusion matrix basically measures wether not a model predicted something as 'true' or 'false' and its actual value. So lets say the model had more false positives than another, that means that model was unable to distinguish wether or not something was actually true. And in the case where it has more false negatives, it has the opposite problem. So the confusion matrix just makes this process easier to read.

The bias comes from errors occurs from underfitting the data. Its like if a you have a chatbot and every time you get a nonsensical response from it, the data is probably underfitted. It is not able to correctly pick up the patterns in the data. Variance is the opposite, it comes from overfitting so if you say put in a chatbot 'whats the best pasta recipe', it will directly output all of its training data for 'whats the best pasta recipe'. It is seriously overfitted, it is trained so much that it even is trained on the background noise of the data, so nonsense is eventually trained into it as not all data equally captures the task at hand. So its clear that in order to get one, you give up the other, the ideal balance being a mix of both. It seems that for my data, the best model was the random forest having this ideal mix since it has a greater t statisic than any other model by far. Ada boost had the weakest performance, it most likely had a very high variance since it failed its p test in comparison to the other models. Gradient boost has medium to high variance. I am saying variance since all these models preformed the same on the training data and if these models did very well on the training data and so it can't have a high bias since it fits very well with the training data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
import seaborn as sns
#
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, f1_score
import numpy as np
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from scipy.stats import ttest_rel
from sklearn.metrics import f1_score
data = pd.read_csv("/content/obesity-prediction-dataset/ObesityDataSet_raw_and_data_sinthetic.csv")
data.head
print(data.describe())
column_names = data.columns
print(column_names)
print(data.isnull().sum())
print(data.isna().sum()[lambda x: x > 0])
label_encoders = {}
for col in data.select_dtypes(include='object').columns.drop('NObeyesdad'):
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
print(data['NObeyesdad'].unique())
data['NObeyesdad'] = data['NObeyesdad'].map({'Insufficient_Weight': 0, 'Normal_Weight': 0, 'Overweight_Level_I': 0, 'Overweight_Level_II': 0, 'Obesity_Type_I': 1, 'Obesity_Type_II': 1, 'Obesity_Type_III': 1})
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
X = data.drop("NObeyesdad", axis=1)
Y = data["NObeyesdad"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=42)
model = DecisionTreeClassifier(random_state=42)
depths = [2, 4, None]
for i in depths:
    model = DecisionTreeClassifier(max_depth = i, random_state=42)
    model.fit(X_train, Y_train)
    plt.figure(figsize=(12, 6))
    plot_tree(model,  feature_names=X.columns, class_names = ['Not_Overweight', 'Overweight'], filled=True)
    plt.title("Decision Tree")
    plt.show()
acc = model.score(X_test, Y_test)
CopXOne = X_test.copy()
CopXTwo = X_test.copy()
CopXOne['Age'] = np.random.permutation(CopXOne['Age'])
CopXTwo['FCVC'] = np.random.permutation(CopXTwo['FCVC'])
accOne = acc - model.score(CopXOne, Y_test)
accTwo = acc - model.score(CopXTwo, Y_test)
print("Age drop: ", accOne)
print("FCVC drop: ", accTwo)
folds = [5,10,15]
models = {
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=40),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=42),
    "GradientBoost": GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42),
}
#orginally I was going to use KFold but later found out that StratifiedKFold was more consistent
#once I averaged it out
fOneALT = {name: [] for name in models}
for m in models:
    mods = models[m]
    for k in folds:
        kf = StratifiedKFold(n_splits=k, shuffle=True, random_state=40)
        acc = []
        precsion = []
        fOne = []
        #the actual spliting of the test occurs here
        for training, test in kf.split(X,Y):
             X_train = X.iloc[training]
             X_test = X.iloc[test]
             Y_train = Y.iloc[training]
             Y_test = Y.iloc[test]
             mods.fit(X_train,Y_train)
             Y_prediction = mods.predict(X_test)
             acc.append(accuracy_score(Y_test, Y_prediction))
             precsion.append(precision_score(Y_test, Y_prediction))
             fOne.append(f1_score(Y_test, Y_prediction))
             fOneALT[m].append(f1_score(Y_test, Y_prediction))
        print("For model", m)
        print("For ", k,"fold, ")
        print("accuracy: ", np.mean(acc))
        print("precsion: ", np.mean(precsion))
        print("fOne: ", np.mean(fOne))
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=40)
for train, test in kf.split(X, Y):
    X_train = X.iloc[training]
    X_test = X.iloc[test]
    Y_train = Y.iloc[training]
    Y_test = Y.iloc[test]
    break
for m in models:
    mods = models[m]
    mods.fit(X_train, Y_train)
    Y_prediction = mods.predict(X_test)
    cm = confusion_matrix(Y_test, Y_prediction)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    print(m);
    plt.show()
tStat, pVal = ttest_rel(fOneALT['RandomForest'], fOneALT['AdaBoost'])
print("RandomForest vs AdaBoost")
print("tStat ", tStat)
print("pVal ",pVal)
tStat, pVal = ttest_rel(fOneALT['RandomForest'], fOneALT['GradientBoost'])
print("RandomForest vs GradientBoost")
print("tStat ", tStat)
print("pVal ",pVal)
tStat, pVal = ttest_rel(fOneALT['AdaBoost'], fOneALT['GradientBoost'])
print("AdaBoost vs GradientBoost")
print("tStat ", tStat)
print("pVal ",pVal)
```
